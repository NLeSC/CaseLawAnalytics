```python
from lxml import etree
import re
import os
import sqlite3
```

```python
conn = sqlite3.connect('rechtspraak.db')
c = conn.cursor()
```

```python
rows = c.execute('''SELECT u.id, m.summary, u.text 
                    from uitspraken u
                    join UITSPRAKEN_META m on u.id = m.id 
                    limit 10''')
```

```python
res = rows.fetchall()
```

```python
root = etree.Element("root")
add = etree.SubElement(root, "add")
```

```python
for id0, abstract, content in res: 
    doc = etree.SubElement(add, "doc")
    field_id = etree.SubElement(doc, "field", name="id")
    field_id.text = "http://deeplink.rechtspraak.nl/uitspraak?id="+id0
    field_id = etree.SubElement(doc, "field", name="abstract")
    field_id.text = abstract
    field_id = etree.SubElement(doc, "field", name="content")
    field_id.text = content
```

```python
print(str(etree.tostring(add)))
```

```python
fn = '../../../solr/example_data/data.xml'
with open(fn, 'w') as out:
    out.write(etree.tounicode(add, pretty_print=True))
```

```python
c.close()
```
